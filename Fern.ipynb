# Una felce frattale
È ora il turno di un altro classicone: la felce.

Come prima cosa, importiamo un po' di librerie e impostiamo qualche parametro…

```python
from matplotlib import pyplot as plt
import numpy as np
import random
```

Ora definiamo la funzione che fa il calcolo vero e proprio; si limita a estrarre un numero tra 0 e 1 e, in base al risultato, scegliere e applicare alle coordinate che le vengono passate una delle trasformazioni che compongono la felce. Al contrario del triangolo di Sierpinski, qui le trasformazioni non hanno tutte lo stesso "peso".

```python
def fern(x, y):
    prob = random.random()
    if prob < 0.01:
        return 0, 0.16 * y, 1
    elif prob < 0.86:
        return 0.85 * x + 0.04 * y, -0.04 * x + 0.85 * y + 1.6, 2
    elif prob < 0.93:
        return 0.2 * x - 0.26 * y, 0.23 * x + 0.22 *y + 1.6, 3
    else:
        return -0.15 * x + 0.28 * y, 0.26* x + 0.24 * y + 0.44, 4
```

Ora inizializziamo il necessario, calcoliamo un po' di iterazioni e riempiano due array con i risultati (coordinate x, coordinate y e "colore")…

```python
iterations = 1000000
nextx, nexty = 0, 0
x, y, c = [0], [0], [0]
    
for _ in range(iterations):
    nextx, nexty, color = fern(nextx, nexty)
    x.append(nextx)                         
    y.append(nexty)  
    c.append(color)
```

Per finire, grafichiamo i punti ottenuti come scatter plot:

```python
plt.figure(figsize = (9,12))
plt.gca().set_facecolor("black")
plt.scatter(x, y, s = 0.01, c = c,  marker='o')
plt.show()
```
