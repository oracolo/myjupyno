# Il triangolo di Sierpinski (e amici)
Proviamo ora a generare un frattale piuttosto celebre; come vedremo, la cosa è abbastanza veloce e indolore.

Come prima cosa, importiamo un po' di librerie e impostiamo qualche parametro…

```python
from matplotlib import pyplot as plt
from random import randrange
```

Ora definiamo la funzione che fa il calcolo vero e proprio; si limita a estrarre un numero da 1 a 3 e, in base al risultato, scegliere e applicare alle coordinate che le vengono passate una delle trasformazioni che compongono il triangolo.

```python
def sierpinski(x, y):
    prob = randrange(3)
    if prob == 1:
        return x/2, y/2, 1
    elif prob == 2:
        return x/2 + 0.5, y/2, 2
    else:
        return x/2, y/2 + 0.5, 3
```

Ora inizializziamo il necessario, calcoliamo un po' di iterazioni e riempiamo due array con i risultati (coordinate x, coordinate y e "colore")…

```python
iterations = 1000000
nextx, nexty = 0, 0
x, y, c = [0], [0], [0]
    
for _ in range(iterations):
    nextx, nexty, color = sierpinski(nextx, nexty)
    x.append(nextx)                         
    y.append(nexty)  
    c.append(color)
```

Per finire, grafichiamo i punti ottenuti come scatter plot:

```python
plt.figure(figsize=(12, 12))
plt.gca().set_facecolor("black")
plt.scatter(x, y, s = 0.01, c = c,  marker='o', cmap = "plasma")
plt.show()
```
