# Il triangolo di Sierpinski (e amici)
Proviamo ora a generare un frattale piuttosto celebre; come vedremo, la cosa è abbastanza veloce e indolore.

Come prima cosa, importiamo un po' di librerie e impostiamo qualche parametro…

```python
from itertools import islice
from matplotlib import pyplot as plt
from random import randrange
```

Ora definiamo la funzione che fa il calcolo vero e proprio; si limita a estrarre un numero da 0 a 2 e, in base al risultato, scegliere e applicare alle coordinate che le vengono passate una delle trasformazioni che compongono il triangolo.

```python
def sierpinski(x=0, y=0):
    while True:
        x /= 2
        y /= 2
        roll = randrange(3)
        if roll == 1:  # with Python 3.10, we could use the new "match" construct
            x += 0.5
        elif roll == 2:
            y += 0.5
        yield x, y, roll
```

Ora inizializziamo il necessario, calcoliamo un po' di iterazioni e riempiano due array con i risultati (coordinate x, coordinate y e "colore")…

```python
iterations = 50_000

x, y, c = zip(*islice(sierpinski(), iterations))
```

Per finire, grafichiamo i punti ottenuti come scatter plot:

```python
plt.figure(figsize=(12, 12))
plt.gca().set_facecolor('black')
plt.scatter(x, y, s=0.01, c=c,  marker='o', cmap='plasma')
plt.show()
```
