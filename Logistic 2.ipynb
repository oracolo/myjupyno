```python
import matplotlib.pyplot as plt
import numpy as np
import math

def logistic(x, r, iter, filter):
    for i in range(iter):
        x = r * x * (1-x)
        if i == 0:
            values = np.array([x])
        elif i>filter:
            values = np.append(values, [x])   
    return values

for r in np.arange(2, 4, 0.005):
    if r == 2:
        y = logistic(0.5, r, 5000, 100)
        x = np.full(4900, fill_value=r)
    else:
        y = np.append(y, logistic(0.5, r, 5000, 100))
        x = np.append(x, np.full(4900, fill_value=r))
        
plt.scatter(x, y, s=(0.01)**2, marker='.')
plt.show()
```
